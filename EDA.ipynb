### Ejercicio uno: realice un analisis de valores nulos

En caso de existir valores nulos elimine esos registros

### Castiando el string a Date

```python
vdf['date'] = pd.to_datetime(vdf['date'])
```

### Creando dos nuevas columnas con year and month values from date

```python
vdf['Date Month'] = vdf['date'].dt.month
vdf['Date Year'] = vdf['date'].dt.year
```

```python
vdf.dtypes
```

```python
vdf.head(3)
```

```python
vdf.loc[ : , "Date Year" ].drop_duplicates()
```

```python
result = pd.DataFrame()
```

```python
for year_ in vdf.loc[ : , "Date Year" ].drop_duplicates():
    vdf_year = vdf.loc[ vdf['Date Year'] == year_ , : ]    
    for column_ in vdf_year.columns:        
        percentage =  np.sum(pd.isna( vdf_year.loc[ : , column_ ]  )) / vdf_year.shape[0]
        dfrow = pd.DataFrame({
            'year': [ year_ ]
            , 'column' : [ column_ ]
            , 'percentage.of.nans' : [ percentage ]
        })
        result = pd.concat([result, dfrow])
```

```python
result
```

### Ejercicio seis: muestre por category,item,variety que tanto % ha aumentado su precio

Ejemplo: si para las manzanas gala el primer precio que se registro fue 2.00 y el ultimo precio que se registro es 4.00 este aumento un 100%

```python
pd.DataFrame({
    "category": ["fruta","fruta"]
    ,"item" : ['Manzana','Manzana']
    , "variety" : ['Gala','egremont_russet']
    , "percentage-change-between-first-and-last-price" : [200,150]       
})
```

```python
vdfE6 = vdf.copy()
```

```python
vdfE6.dtypes
```

```python
x = vdfE6.sort_values(["item","Date Year"] , ascending = False )
x["rank"] = x.groupby(["item"])["Date Year"].rank( method = 'first' , ascending = False )
x = x.loc[ x["rank"] == 1.0 , : ].reset_index(drop=True)
```

```python
y = vdfE6.sort_values(["item","Date Year"] , ascending = True )
y["rank"] = y.groupby(["item"])["Date Year"].rank( method = 'first' , ascending = True )
y = y.loc[ y["rank"] == 1.0 , : ].sort_values(["item"] , ascending = False ).reset_index(drop=True)
```

```python
percentage_price =(x['price'] / y['price'].sum()) * 100
```

```python
vdfE6["percentage_price"] = percentage_price
```

```python
vdfE6 = vdfE6.loc[: , ["category", "item", "variety", "percentage_price"]]
```
